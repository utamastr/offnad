```python
import numpy as np

def quatmultiply(q1, q2):
    x1, y1, z1, s1 = q1
    x2, y2, z2, s2 = q2
    x = s1 * x2 + x1 * s2 + y1 * z2 - z1 * y2
    y = s1 * y2 + y1 * s2 + z1 * x2 - x1 * z2
    z = s1 * z2 + z1 * s2 + x1 * y2 - y1 * x2
    s = s1 * s2 - x1 * x2 - y1 * y2 - z1 * z2
    magnitude = np.sqrt(x**2 + y**2 + z**2 + s**2)
    if magnitude > 0:
        x /= magnitude
        y /= magnitude
        z /= magnitude
        s /= magnitude
    return [x, y, z, s]

def quatconj(quaternion):
    quaternion_ = [-quaternion[0], -quaternion[1], -quaternion[2], quaternion[3]]
    return quaternion_

def quatrotate(quaternion, vector):
    # new_vector = quatmultiply(quatconj(quaternion), quatmultiply([vector[0], vector[1], vector[2], 0], quaternion))
    new_vector = quatmultiply(quatmultiply(quaternion, [vector[0], vector[1], vector[2], 0]), quatconj(quaternion))
    return new_vector[0:3]

def ypr2q(yaw, pitch, roll):
    q_z = [0, 0, np.sin(yaw / 2), np.cos(yaw / 2)]
    q_y = [0, np.sin(pitch / 2), 0, np.cos(pitch / 2)]
    q_x = [np.sin(roll / 2), 0, 0, np.cos(roll / 2)]
    q = quatmultiply(q_z, quatmultiply(q_y, q_x))
    magnitude = np.sqrt(q[0]**2 + q[1]**2 + q[2]**2 + q[3]**2)
    q /= magnitude
    return q

import numpy as np

def quaternion2ypr(q):
    magnitude = np.sqrt(q[0]**2 + q[1]**2 + q[2]**2 + q[3]**2)
    q /= magnitude
    x, y, z, s = q

    # Yaw (Z-axis rotation)
    siny_cosp = 2 * (s * z + x * y)
    cosy_cosp = 1 - 2 * (y * y + z * z)
    yaw = np.arctan2(siny_cosp, cosy_cosp)

    # Pitch (Y-axis rotation)
    sinp = 2 * (s * y - z * x)
    if np.abs(sinp) >= 1:
        pitch = np.copysign(np.pi / 2, sinp)  # Use 90 degrees if out of range
    else:
        pitch = np.arcsin(sinp)

    # Roll (X-axis rotation)
    sinr_cosp = 2 * (s * x + y * z)
    cosr_cosp = 1 - 2 * (x * x + y * y)
    roll = np.arctan2(sinr_cosp, cosr_cosp)

    yaw = np.degrees(yaw)
    pitch = np.degrees(pitch)
    roll = np.degrees(roll)
    return yaw, pitch, roll

def dcm2quat(R):
    trace = np.trace(R)
    
    if trace > 0:
        s = 0.5 / np.sqrt(trace + 1.0)
        w = 0.25 / s
        x = (R[2, 1] - R[1, 2]) * s
        y = (R[0, 2] - R[2, 0]) * s
        z = (R[1, 0] - R[0, 1]) * s
    else:
        if R[0, 0] > R[1, 1] and R[0, 0] > R[2, 2]:
            s = 2.0 * np.sqrt(1.0 + R[0, 0] - R[1, 1] - R[2, 2])
            w = (R[2, 1] - R[1, 2]) / s
            x = 0.25 * s
            y = (R[0, 1] + R[1, 0]) / s
            z = (R[0, 2] + R[2, 0]) / s
        elif R[1, 1] > R[2, 2]:
            s = 2.0 * np.sqrt(1.0 + R[1, 1] - R[0, 0] - R[2, 2])
            w = (R[0, 2] - R[2, 0]) / s
            x = (R[0, 1] + R[1, 0]) / s
            y = 0.25 * s
            z = (R[1, 2] + R[2, 1]) / s
        else:
            s = 2.0 * np.sqrt(1.0 + R[2, 2] - R[0, 0] - R[1, 1])
            w = (R[1, 0] - R[0, 1]) / s
            x = (R[0, 2] + R[2, 0]) / s
            y = (R[1, 2] + R[2, 1]) / s
            z = 0.25 * s
    return np.array([x, y, z, w])

from scipy.spatial.transform import Rotation as R

eph = load('de421.bsp')
earth, sun = eph['earth'], eph['sun']
ts=load.timescale()

def ypr_from_eci(t_s, q, lon, lat, v):
    # ts=load.timescale()
    t = datetime.strptime(t_s, '%Y-%m-%d %H:%M:%S.%f')
    tn = ts.utc(t.year, t.month, t.day, t.hour, t.minute, t.second)
    gmst_rad = tn.gmst/24*np.deg2rad(360)

    q10 = np.cos((np.deg2rad(lon) + gmst_rad)/2)
    q11 = 0
    q12 = 0
    q13 = np.sin((np.deg2rad(lon) + gmst_rad)/2)
    # r1 = R.from_quat([q11, q12, q13, q10])
    # M1 = r1.as_matrix()
    q1 = [q11, q12, q13, q10]

    q20 = np.cos((-1.570796326794897-np.deg2rad(lat))/2)
    q21 = 0
    q22 = np.sin((-1.570796326794897-np.deg2rad(lat))/2)
    q23 = 0
    # r2 = R.from_quat([q21, q22, q23, q20])
    # M2 = r2.as_matrix()
    q2 = [q21, q22, q23, q20]

    # M_ned_eci = np.linalg.inv(np.dot(M1, M2))
    q_eci_ned = quatmultiply(q1, q2)
    q_ned_eci = quatconj(q_eci_ned)

    # r_eci_sat = R.from_quat(q)
    # M_eci_sat = np.dot(r_eci_sat.as_matrix(), [[-1, 0, 0], [0, -1, 0], [0, 0, 1]])
    # M_ned_sat = np.dot(M_ned_eci, M_eci_sat)
    # q_eci_sat = quatmultiply(q, [0, 0, 1, 0])
    q_eci_sat = q
    q_ned_sat = quatmultiply(q_ned_eci, q_eci_sat)
    
    norm_v = np.linalg.norm(v, axis=0)
    v_eci_sat = v/norm_v
    # v_ned_sat = np.dot(M_ned_eci, v_eci_sat)
    v_ned_sat = quatrotate(q_ned_eci, v_eci_sat)
    az_v = np.arccos(v_ned_sat[0])
    
    # M_nad_ned = np.array([[np.cos(az_v), -np.sin(az_v), 0],
    #                       [np.sin(az_v), np.cos(az_v), 0],
    #                       [0,0,1]])
    # M_nad_sat = np.dot(M_nad_ned, M_ned_sat)
    # r_nad_sat = R.from_matrix(M_nad_sat)
    # q_nad_sat = r_nad_sat.as_quat()
    # ypr = r_nad_sat.as_euler('zyx', degrees=True)
    q_nad_ned = ypr2q(az_v, 0, 0)
    q_nad_sat = quatmultiply(q_nad_ned, q_ned_sat)
    y, p, r = quaternion2ypr(q_nad_sat)
    ypr = [y, p, r]
    
    return q_nad_sat, ypr

def sunmag_ad(lon, lat, alt, mag_x, mag_y, mag_z, sun_x, sun_y, sun_z, t_s, v):
    t = datetime.strptime(t_s, '%Y-%m-%d %H:%M:%S.%f')
    earth, sun = eph['earth'], eph['sun']
    tn = ts.utc(t.year, t.month, t.day, t.hour, t.minute, t.second)
    gmst_rad = tn.gmst/24*np.deg2rad(360)

    q10 = np.cos((np.deg2rad(lon) + gmst_rad)/2)
    q11 = 0
    q12 = 0
    q13 = np.sin((np.deg2rad(lon) + gmst_rad)/2)
    q1 = [q11, q12, q13, q10]
    q20 = np.cos((-1.570796326794897-np.deg2rad(lat))/2)
    q21 = 0
    q22 = np.sin((-1.570796326794897-np.deg2rad(lat))/2)
    q23 = 0
    q2 = [q21, q22, q23, q20]
    q_eci_ned = quatmultiply(q1, q2)
    q_ned_eci = quatconj(q_eci_ned)

    a, b, c = ppigrf.igrf(lon, lat, alt, t)
    BN = b[0][0]
    BE = a[0][0]
    BD = -c[0][0]

    norm_B = np.linalg.norm([BN, BE, BD], axis=0)
    bn = np.array(BN/norm_B)
    be = np.array(BE/norm_B)
    bd = np.array(BD/norm_B)

    # tn = ts.utc(t.year, t.month, t.day, t.hour, t.minute, t.second)
    satellite_location = Topos(latitude_degrees=lat, longitude_degrees=lon, elevation_m=alt*1000)
    sun_pos = (earth + satellite_location).at(tn).observe(sun).apparent()
    el, az, d = sun_pos.altaz()
    SN = np.array(np.cos(np.deg2rad(el.degrees))*np.cos(np.deg2rad(az.degrees)))
    SE = np.array(np.cos(np.deg2rad(el.degrees))*np.sin(np.deg2rad(az.degrees)))
    SD = np.array(-np.sin(np.deg2rad(el.degrees)))

    SxB = np.cross([SN, SE, SD], [bn, be, bd])
    BxSxB = np.cross([bn, be, bd], SxB)
    M_ned_sm = np.array([[bn, be, bd], SxB, BxSxB])
    q_ned_sm = dcm2quat(M_ned_sm)

    A = np.array([[-1.01273467059234, -0.0162932050559462, -0.0106220892014218],
                  [-0.0162932050559462, -0.983015277270010, 0.0120503962203343],
                  [0.0106220892014218, -0.0120503962203343, 1.02801262026923]])
    C = np.array([[-0.0047, -1, -0.0015],
                  [0.0095, -0.0015, 1],
                  [-0.9999, 0.0047, 0.0096]])
    b = np.array([3972.00, -6166.57, 7592.22])
    mag = [mag_x, mag_y, mag_z]
    sun = [sun_x, sun_y, sun_z]
    mag_body = np.dot(A, np.dot(C, mag) - b)
    mag_norm = mag_body/np.linalg.norm(mag_body)

    sunxmag = np.cross(sun, mag_norm)
    magxsunxmag = np.cross(mag_norm, sunxmag)
    M_sat_sm = np.array([mag_norm, sunxmag, magxsunxmag])
    q_sat_sm = dcm2quat(M_sat_sm)
    q_ned_sat = quatmultiply(quatconj(q_ned_sm), q_sat_sm)
        
    norm_v = np.linalg.norm(v, axis=0)
    v_eci_sat = v/norm_v
    # v_ned_sat = np.dot(M_ned_eci, v_eci_sat)
    v_ned_sat = quatrotate(q_ned_eci, v_eci_sat)
    az_v = np.arccos(v_ned_sat[0])
    
    q_nad_ned = ypr2q(az_v, 0, 0)
    q_nad_sat = quatmultiply(q_nad_ned, q_ned_sat)
    y, p, r = quaternion2ypr(q_nad_sat)
    ypr = [y, p, r]

    return q, ypr
```
